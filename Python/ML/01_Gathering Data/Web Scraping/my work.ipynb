```python
for tag in soup.find_all('div', class_='style__width-100p___2woP5 style__flex-row___m8FHw'):
    lines = list(tag.stripped_strings)

    name = lines[0] if lines else None
    price = None
    prescription_required = False
    company = None
    salt = None

    for i, line in enumerate(lines):
        # Find '₹' followed by the price
        if line.strip() == '₹' and i + 1 < len(lines):
            price = lines[i + 1].strip()

        elif 'Prescription Required' in line:
            prescription_required = True

        elif re.search(r'(Ltd|Laboratories|Pharma|Cipla|Pfizer)', line, re.IGNORECASE):
            company = line.strip()

        elif re.search(r'\(.*?(mg|mcg|ml).*?\)', line):
            salt = line.strip()

    print("Name:", name)
    print("Price: Rs.", price,)
    print("Prescription Required:", prescription_required)
    print("Company:", company)
    print("Salt:", salt)
    print("=" * 40)
```

```python
all_data = []
BASE_URL = "https://www.1mg.com/drugs-all-medicines?page="
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Referer': 'https://www.google.com',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
    'Connection': 'keep-alive',
}


for page in range(1, 345):  # from 1 to 344
    print(f"Scraping page {page}...")

    url = f"{BASE_URL}{page}"
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    for tag in soup.find_all('div', class_='style__width-100p___2woP5 style__flex-row___m8FHw'):
        lines = list(tag.stripped_strings)

        if not lines:
            continue

        name = lines[0]
        price = None
        prescription_required = False
        company = None
        salt = None

        for i, line in enumerate(lines):
            if line.strip() == '₹' and i + 1 < len(lines):
                price = lines[i + 1].strip()

            elif 'Prescription Required' in line:
                prescription_required = True

            elif re.search(r'(Ltd|Laboratories|Pharma|Cipla|Pfizer)', line, re.IGNORECASE):
                company = line.strip()

            elif re.search(r'\(.*?(mg|mcg|ml).*?\)', line):
                salt = line.strip()

        all_data.append({
            'Name': name,
            'Price': price,
            'Prescription Required': prescription_required,
            'Company': company,
            'Salt': salt
        })

    time.sleep(1)  # Be polite — don't hammer the server
```

```python
all_data
```

```python
with open('medicine_data.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['Name', 'Price', 'Prescription Required', 'Company', 'Salt'])
    writer.writeheader()
    writer.writerows(all_data)
```
